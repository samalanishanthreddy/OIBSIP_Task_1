## Introduction 

The Iris flower classification problem involves categorizing iris flowers into one of three species (classes) based on their botanical features. The dataset used for this classification is called the Iris dataset, which was introduced by the British biologist and statistician Ronald A. Fisher in 1936. It is a widely used dataset in machine learning and statistics.

## Aim of this project

Given the measurements of sepal length, sepal width, petal length, and petal width of iris flowers, the task is to classify the flower to which Iris species (Setosa, Versicolor, or Virginica) it belongs to, by trainig machine learning model that will classify them.
